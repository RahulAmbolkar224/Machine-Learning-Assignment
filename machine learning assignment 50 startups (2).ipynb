# R & D Spend and Profit Linear Regression

```python
x= df['R&D Spend'].values
y= df['Profit'].values
```

```python
x = x.reshape(-1,1)
x
```

```python
y = y.reshape(-1,1)
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin= LinearRegression()
```

```python
lin.fit(x,y)
```

```python
plt.scatter(x,y,color = 'blue')
plt.plot(x,lin.predict(y),color = 'red')
```
